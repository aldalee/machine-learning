```python
# 设置颜色主题
antV = ['#1890FF', '#2FC25B', '#FACC14', '#223273', '#8543E0', '#13C2C2', '#3436c7', '#F04864'] 
```

```python
# 绘制  Violinplot
f, axes = plt.subplots(2, 2, figsize=(8, 8), sharex=True)
sns.despine(left=True)

sns.violinplot(x='target', y='sepal_length', data=iris, palette=antV, ax=axes[0, 0])
sns.violinplot(x='target', y='sepal_width', data=iris, palette=antV, ax=axes[0, 1])
sns.violinplot(x='target', y='petal_length', data=iris, palette=antV, ax=axes[1, 0])
sns.violinplot(x='target', y='petal_width', data=iris, palette=antV, ax=axes[1, 1])

plt.show()
```
